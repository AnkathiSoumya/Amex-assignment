## 4. Outlier Detection and Removal

```python
df.dtypes
```

```python
num_cols = [
    'no_of_adults', 'no_of_children', 'no_of_weekend_nights', 'no_of_week_nights',
    'lead_time', 'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled',
    'avg_price_per_room', 'no_of_special_requests'
]
```

```python
plt.figure(figsize=(15, 20))
for i, col in enumerate(num_cols, 1):
    plt.subplot(5, 2, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
def detect_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    print(f"{col}: {len(outliers)} outliers")
    return outliers.index
```

```python
outlier_indices = set()
for col in num_cols:
    outlier_idx = detect_outliers_iqr(df, col)
    outlier_indices.update(outlier_idx)

print(f"Total unique outlier rows: {len(outlier_indices)}")
```

## 5. Splitting into Training and Testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['booking_status'])
y = df['booking_status']
```

**3. Splitting dataset 70% Train and 30% Test dataset**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print("Training Data Shape: ", X_train.shape)
print("Testing Data Shape: ", X_test.shape)
```

## 6. Model Building - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
```

```python
lg = LogisticRegression(max_iter=1000)
lg.fit(X_train, y_train)
```

```python
y_pred = lg.predict(X_test)
```

```python
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
results.head()
```

```python
results['Actual'].plot(kind='hist', bins=20, title='Actual')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
results['Predicted'].plot(kind='hist', bins=20, title='Predicted')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

## 7. Model Analysis

**4. Model coefficients**

```python
coefficients = lg.coef_[0]
features = X_train.columns

coef_df = pd.DataFrame({
    'Feature': features,
    'Coefficient': coefficients
})

coef_df['Abs_Coefficient'] = coef_df['Coefficient'].abs()
coef_df_sorted = coef_df.sort_values(by='Abs_Coefficient', ascending=False)

print(coef_df_sorted[['Feature', 'Coefficient']])
```

**5. Model Accuracy**

```python
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
```

```python
print(f"Accuracy: {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall: {rec:.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```



### **Recommendations to Hotel Management Based on Predictive Model and Data Analysis**

Based on the analysis of booking data and the interpretation of the logistic regression model coefficients, the following recommendations are made to help reduce booking cancellations:

---

#### 1. Encourage Shorter Lead Times

**Observation**: The `lead_time` feature has a strong positive coefficient, indicating that longer lead times are associated with a higher probability of cancellations.

**Recommendation**:

* Promote last-minute or near-date bookings through discounts and marketing campaigns.
* Implement stricter cancellation policies for bookings made well in advance, while offering more flexible terms for near-date bookings.

---

#### 2. Promote Repeat Guest Programs

**Observation**: The `repeated_guest` feature has a strong negative coefficient, showing that repeat guests are significantly less likely to cancel.

**Recommendation**:

* Develop loyalty programs or exclusive benefits to encourage guests to book repeatedly.
* Track guest history and offer targeted promotions to returning customers.

---

#### 3. Encourage Special Requests and Parking Reservations

**Observation**: Features like `no_of_special_requests` and `required_car_parking_space` have strong negative coefficients, meaning guests making such requests are less likely to cancel.

**Recommendation**:

* Encourage guests to specify special requests or book parking spaces during the reservation process.
* Highlight these options prominently during booking to foster commitment.

---

#### 4. Focus on Corporate and Offline Booking Channels

**Observation**: Bookings from the `corporate` and `offline` market segments show a reduced likelihood of cancellation.

**Recommendation**:

* Strengthen partnerships with corporate clients and travel agencies.
* Offer special packages or rates for corporate and offline bookings.

---

#### 5. Monitor Online Booking Behavior

**Observation**: The `online` market segment has a slightly positive coefficient, indicating a mild association with cancellations.

**Recommendation**:

* Introduce booking confirmation steps such as email verification or deposits.
* Provide incentives for guests to opt for non-refundable options when booking online.

---

#### 6. Optimize Pricing Strategy

**Observation**: The `avg_price_per_room` feature has a positive coefficient, suggesting that higher room prices may lead to increased cancellations.

**Recommendation**:

* Implement dynamic pricing to balance competitiveness with profitability.
* Bundle services with higher-priced rooms to increase perceived value.
* Offer discounted rates for non-cancellable bookings to reduce price-sensitive cancellations.

---

#### 7. Analyze Room Type Preferences

**Observation**: Certain room types, such as `Room_Type 6` and `Room_Type 7`, are associated with a lower likelihood of cancellation.

**Recommendation**:

* Investigate the characteristics of these room types and promote them more actively.
* Consider adopting similar features in other room types to reduce cancellations across the board.
